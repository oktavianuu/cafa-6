## Reading test data input, the file is in .pasta file
### This is the seq I am trying to predict

```python
from Bio import SeqIO
```

```python
xtrain_data = '/home/okta/Documents/cafa-6/data/Train/train_sequences.fasta'
```

```python
# For a single FASTA record:
try:
    record = SeqIO.read(xtrain_data, "fasta")
    print(f"Sequence ID: {record.id}")
    print(f"Sequence: {record.seq}")
    print(f"Sequence Length: {len(record.seq)}")
except ValueError:
    print(f"The file '{xtrain_data}' contains multiple records. Use SeqIO.parse for multiple records.")
```

#### I use this because the file contains multuple records instead of single records

```python
# For multiple FASTA records:
records = list(SeqIO.parse(xtrain_data, "fasta"))

for i, record in enumerate(records):
    print(f"\nRecord {i+1}:")
    print(f"Sequence ID: {record.id}")
    print(f"Sequence: {record.seq[:50]}...") # Print first 50 characters
    print(f"Sequence Length: {len(record.seq)}")
```
